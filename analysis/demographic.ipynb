```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
import seaborn as sns
from collections import defaultdict
sns.set_theme()

participants = [j for i, j in demographic.items() if True]
n = len(participants)


gender_counts = defaultdict(int)
for p in participants:
    gender_counts[f"gender:{p['gender']}"] += 1
    gender_counts[f"experience:{p['experience']}"] += 1
    gender_counts[f"age:{p['age']}"] += 1
    gender_counts[f"education:{p['education']}"] += 1

for k in sorted(gender_counts):
    print(f"{k}: {gender_counts[k]}")

fields = {
    'Age': 'age',
    'Gender': 'gender',
    'Education': 'education',
    'Experience': 'experience',
}

label_map = {
    '18-25': '18–25', '26-35': '26–35', '36-45': '36–45',
    '46-55': '46–55', '56-65': '56–65', '65+': '65+',
    'male': 'Male', 'female': 'Female', 'non-binary': 'Non-binary',
    'other': 'Other', 'prefer-not-to-say': 'Prefer not to say',
    'high-school': 'High School', 'associate': 'Associate',
    'bachelor': "Bachelor's", 'master': "Master's", 'phd': 'PhD',
    '4-6': '4–6 yrs', '7-10': '7–10 yrs', 'more-than-10': '10+ yrs',
    '5-10': '5–10 yrs', '10+': '10+ yrs',
}


field_order = {
    'Age': ('age', ['18-25', '26-35', '36-45', '46-55', '56-65', '65+', 'prefer-not-to-say']),
    'Gender': ('gender', ['male', 'female', 'non-binary', 'other', 'prefer-not-to-say']),
    'Education': ('education', ['high-school', 'associate', 'bachelor', 'master', 'phd', 'prefer-not-to-say', 'other']),
    'Experience': ('experience', ['less-than-1', '1-3', '4-6', '7-10', 'more-than-10']),
}

results = {}
for display, (key, order) in field_order.items():
    counts = Counter(p[key] for p in participants)
    results[display] = {label_map.get(k, k): round(counts[k] / n * 100, 1) for k in order if counts.get(k, 0) > 0}



fig, ax = plt.subplots(figsize=(9.2, 5))
ax.invert_yaxis()
ax.xaxis.set_visible(False)
ax.set_xlim(0, 100)

colors = plt.cm.tab10.colors

for i, (field, subs) in enumerate(results.items()):
    names = list(subs.keys())
    pcts = list(subs.values())
    left = 0
    for j, (name, pct) in enumerate(zip(names, pcts)):
        rect = ax.barh(field, pct, left=left, height=0.5, color=colors[j])
        if pct > 5:
            ax.bar_label(rect, labels=[f'{name}\n{pct:.0f}%'], label_type='center',
                         color='white', fontsize=7)
        left += pct

ax.xaxis.set_visible(True)
ax.set_xlim(0, 100)
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}%'))
ax.grid(axis='x', linestyle='--', alpha=1)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
import seaborn as sns
from collections import defaultdict
sns.set_theme()

participants = [j for i, j in demographic.items() if i in study_prolific]
n = len(participants)

gender_counts = defaultdict(int)
for p in participants:
    gender_counts[f"gender:{p['gender']}"] += 1
    gender_counts[f"experience:{p['experience']}"] += 1
    gender_counts[f"age:{p['age']}"] += 1
    gender_counts[f"education:{p['education']}"] += 1

for k in sorted(gender_counts):
    print(f"{k}: {gender_counts[k]}")

fields = {
    'Age': 'age',
    'Gender': 'gender',
    'Education': 'education',
    'Experience': 'experience',
}

label_map = {
    '18-25': '18–25', '26-35': '26–35', '36-45': '36–45',
    '46-55': '46–55', '56-65': '56–65', '65+': '65+',
    'male': 'Male', 'female': 'Female', 'non-binary': 'Non-binary',
    'other': 'Other', 'prefer-not-to-say': 'Prefer not to say',
    'high-school': 'High School', 'associate': 'Associate',
    'bachelor': "Bachelor's", 'master': "Master's", 'phd': 'PhD',
    '4-6': '4–6 yrs', '7-10': '7–10 yrs', 'more-than-10': '10+ yrs',
    '5-10': '5–10 yrs', '10+': '10+ yrs',
}


field_order = {
    'Age': ('age', ['18-25', '26-35', '36-45', '46-55', '56-65', '65+', 'prefer-not-to-say']),
    'Gender': ('gender', ['male', 'female', 'non-binary', 'other', 'prefer-not-to-say']),
    'Education': ('education', ['high-school', 'associate', 'bachelor', 'master', 'phd', 'prefer-not-to-say', 'other']),
    'Experience': ('experience', ['less-than-1', '1-3', '4-6', '7-10', 'more-than-10']),
}

results = {}
for display, (key, order) in field_order.items():
    counts = Counter(p[key] for p in participants)
    results[display] = {label_map.get(k, k): round(counts[k] / n * 100, 1) for k in order if counts.get(k, 0) > 0}



fig, ax = plt.subplots(figsize=(9.2, 5))
ax.invert_yaxis()
ax.xaxis.set_visible(False)
ax.set_xlim(0, 100)

colors = plt.cm.tab10.colors

for i, (field, subs) in enumerate(results.items()):
    names = list(subs.keys())
    pcts = list(subs.values())
    left = 0
    for j, (name, pct) in enumerate(zip(names, pcts)):
        rect = ax.barh(field, pct, left=left, height=0.5, color=colors[j])
        if pct > 5:
            ax.bar_label(rect, labels=[f'{name}\n{pct:.0f}%'], label_type='center',
                         color='white', fontsize=7)
        left += pct

ax.xaxis.set_visible(True)
ax.set_xlim(0, 100)
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}%'))
ax.grid(axis='x', linestyle='--', alpha=1)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```
