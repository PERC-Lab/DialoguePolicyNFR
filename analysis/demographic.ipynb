```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

participants = [j for i, j in demographic.items() if i in pilot_uni]
n = len(participants)

fields = {
    'age': 'Age',
    'gender': 'Gender',
    'education': 'Education',
    'experience': 'Experience (yrs)',
}

all_rows = []
for field, label in fields.items():
    counts = Counter(p[field] for p in participants)
    cats = list(counts.keys())
    vals = np.array(list(counts.values()), dtype=float)
    all_rows.append((label, cats, vals / vals.sum() * 100, vals.astype(int)))

fig, ax = plt.subplots(figsize=(9.2, 0.9 * len(all_rows) + 1.2))
ax.invert_yaxis()
ax.set_xlim(0, 100)

# Show x-axis with percentage ticks at the bottom
ax.set_xticks(range(0, 101, 10))
ax.set_xticklabels([f'{x}%' for x in range(0, 101, 10)], fontsize=8)
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(axis='x', length=0, pad=5)

# Remove spines except bottom
for spine in ax.spines.values():
    spine.set_visible(False)
ax.tick_params(left=False)

max_cats = max(len(cats) for _, cats, _, _ in all_rows)
cmap = plt.colormaps['tab10']
category_colors = [cmap(i / max(max_cats - 1, 1)) for i in range(max_cats)]

labels = [label for label, _, _, _ in all_rows]

for row_idx, (label, cats, pcts, raw) in enumerate(all_rows):
    left = 0.0
    for i, (cat, pct, cnt) in enumerate(zip(cats, pcts, raw)):
        color = category_colors[i % len(category_colors)]
        ax.barh(row_idx, pct, left=left, height=0.5,
                color=color, edgecolor='white', linewidth=0.6)

        if pct >= 6:
            r, g, b, _ = color
            text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
            ax.text(left + pct / 2, row_idx,
                    f'{cat}\n{cnt}',
                    ha='center', va='center',
                    fontsize=7, color=text_color, fontweight='semibold')
        left += pct

ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels, fontweight='bold', fontsize=10)

fig.suptitle(f'Participant Demographics (n = {n})',
             fontsize=12, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

#############
lang_counts = Counter(lang for p in participants for lang in p.get('languages', []))

# Top 6 + "Other"
top6 = lang_counts.most_common(8)
top6_langs = {l for l, _ in top6}
other_count = sum(c for l, c in lang_counts.items() if l not in top6_langs)

langs = [l for l, _ in top6] + ['Other']
counts = np.array([c for _, c in top6] + [other_count])
pcts = counts / n * 100

fig, ax = plt.subplots(figsize=(9.2, 2.8))
ax.invert_yaxis()

color = '#4E79A7'

bars = ax.barh(range(len(langs)), pcts, height=0.6,
               color=color, edgecolor='white', linewidth=0.6)

for i, (pct, cnt) in enumerate(zip(pcts, counts)):
    if pct >= 12:
        ax.text(pct / 2, i, f'{cnt} ({pct:.0f}%)',
                ha='center', va='center',
                fontsize=8, color='white', fontweight='semibold')
    else:
        ax.text(pct + 1, i, f'{cnt} ({pct:.0f}%)',
                ha='left', va='center',
                fontsize=8, color='dimgrey', fontweight='semibold')

ax.set_yticks(range(len(langs)))
ax.set_yticklabels(langs, fontsize=10, fontweight='bold')

ax.set_xlim(0, 105)
ax.set_xticks(range(0, 101, 10))
ax.set_xticklabels([f'{x}%' for x in range(0, 101, 10)], fontsize=8)
ax.tick_params(axis='x', length=0, pad=5)
ax.margins(y=0.08)

for spine in ax.spines.values():
    spine.set_visible(False)
ax.tick_params(left=False)

fig.suptitle(f'Programming Languages (n = {n})',
             fontsize=12, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

participants = [j for i, j in demographic.items() if True]
n = len(participants)

fields = {
    'age': 'Age',
    'gender': 'Gender',
    'education': 'Education',
    'experience': 'Experience (yrs)',
}

all_rows = []
for field, label in fields.items():
    counts = Counter(p[field] for p in participants)
    cats = list(counts.keys())
    vals = np.array(list(counts.values()), dtype=float)
    all_rows.append((label, cats, vals / vals.sum() * 100, vals.astype(int)))

fig, ax = plt.subplots(figsize=(9.2, 0.9 * len(all_rows) + 1.2))
ax.invert_yaxis()
ax.set_xlim(0, 100)

# Show x-axis with percentage ticks at the bottom
ax.set_xticks(range(0, 101, 10))
ax.set_xticklabels([f'{x}%' for x in range(0, 101, 10)], fontsize=8)
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(axis='x', length=0, pad=5)

# Remove spines except bottom
for spine in ax.spines.values():
    spine.set_visible(False)
ax.tick_params(left=False)

max_cats = max(len(cats) for _, cats, _, _ in all_rows)
cmap = plt.colormaps['tab10']
category_colors = [cmap(i / max(max_cats - 1, 1)) for i in range(max_cats)]

labels = [label for label, _, _, _ in all_rows]

for row_idx, (label, cats, pcts, raw) in enumerate(all_rows):
    left = 0.0
    for i, (cat, pct, cnt) in enumerate(zip(cats, pcts, raw)):
        color = category_colors[i % len(category_colors)]
        ax.barh(row_idx, pct, left=left, height=0.5,
                color=color, edgecolor='white', linewidth=0.6)

        if pct >= 6:
            r, g, b, _ = color
            text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
            ax.text(left + pct / 2, row_idx,
                    f'{cat}\n{cnt}',
                    ha='center', va='center',
                    fontsize=7, color=text_color, fontweight='semibold')
        left += pct

ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels, fontweight='bold', fontsize=10)

fig.suptitle(f'Participant Demographics (n = {n})',
             fontsize=12, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()

#############
lang_counts = Counter(lang for p in participants for lang in p.get('languages', []))

# Top 6 + "Other"
top6 = lang_counts.most_common(8)
top6_langs = {l for l, _ in top6}
other_count = sum(c for l, c in lang_counts.items() if l not in top6_langs)

langs = [l for l, _ in top6] + ['Other']
counts = np.array([c for _, c in top6] + [other_count])
pcts = counts / n * 100

fig, ax = plt.subplots(figsize=(9.2, 2.8))
ax.invert_yaxis()

color = '#4E79A7'

bars = ax.barh(range(len(langs)), pcts, height=0.6,
               color=color, edgecolor='white', linewidth=0.6)

for i, (pct, cnt) in enumerate(zip(pcts, counts)):
    if pct >= 12:
        ax.text(pct / 2, i, f'{cnt} ({pct:.0f}%)',
                ha='center', va='center',
                fontsize=8, color='white', fontweight='semibold')
    else:
        ax.text(pct + 1, i, f'{cnt} ({pct:.0f}%)',
                ha='left', va='center',
                fontsize=8, color='dimgrey', fontweight='semibold')

ax.set_yticks(range(len(langs)))
ax.set_yticklabels(langs, fontsize=10, fontweight='bold')

ax.set_xlim(0, 105)
ax.set_xticks(range(0, 101, 10))
ax.set_xticklabels([f'{x}%' for x in range(0, 101, 10)], fontsize=8)
ax.tick_params(axis='x', length=0, pad=5)
ax.margins(y=0.08)

for spine in ax.spines.values():
    spine.set_visible(False)
ax.tick_params(left=False)

fig.suptitle(f'Programming Languages (n = {n})',
             fontsize=12, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
