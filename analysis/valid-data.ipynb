```python
# Count how many participants answered each batch using defaultdict
from collections import defaultdict

batch_counts = defaultdict(int)
for v in valid_participants:
    data = nfr_response[v]
    batch = data[0]["batch"]
    batch_counts[batch] += 1


for batch in sorted(batch_counts):
    print(f"Batch {batch}: {batch_counts[batch]} participants")
```

```python
x = 0
for i,j in conversation.items():
    if i in valid_participants:
        x += len(j)
x
```

```python
# Check if att contains the specified values for each valid participant
values_to_check = {
    "nfr_display_1": "Agree",
    "nfr_display_2": "Agree",
    "nfr_feedback_1": "Disagree",
    "nfr_feedback_2": "4"
}

for i, j in nfr_response.items():
    if i in valid_participants:
        att = j[-1]
        matches = all(att.get(k) == v for k, v in values_to_check.items())
        if not matches:
            print(f"Participant {i}: {matches}")
```
