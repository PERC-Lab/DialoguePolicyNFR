# Dialogue statistics

```python
avg_turns_per_dialogue = []
avg_words_in_prompt = []
max_words_in_prompt = []
avg_words_in_response = []
max_words_in_response = []
avg_words_per_turn = []
max_words_per_turn = []
total_dialogues = []
total_turns = []

for dialogue in conversation.values():
    avg_turns_per_dialogue.append(len(dialogue))

    prompt_word_counts = []
    response_word_counts = []
    turn_word_counts = []

    for turn in dialogue:
        p_words = len(turn["user_message"].split())
        r_words = len(turn["bot_reply"].split())

        prompt_word_counts.append(p_words)
        response_word_counts.append(r_words)
        turn_word_counts.append(p_words + r_words)

    avg_words_in_prompt.append(sum(prompt_word_counts) / len(prompt_word_counts))
    max_words_in_prompt.append(max(prompt_word_counts))
    avg_words_in_response.append(sum(response_word_counts) / len(response_word_counts))
    max_words_in_response.append(max(response_word_counts))
    avg_words_per_turn.append(sum(turn_word_counts) / len(turn_word_counts))
    max_words_per_turn.append(max(turn_word_counts))

total_dialogues = len(conversation)
total_turns = sum(avg_turns_per_dialogue)

print(f"Total Dialogues: {total_dialogues}")
print(f"Total Turns: {total_turns}")
print(f"Avg Turns per Dialogue: {sum(avg_turns_per_dialogue) / len(avg_turns_per_dialogue):.2f}")
print(f"Avg Words in Prompt: {sum(avg_words_in_prompt) / len(avg_words_in_prompt):.2f}")
print(f"Max Words in Prompt: {max(max_words_in_prompt)}")
print(f"Avg Words in Response: {sum(avg_words_in_response) / len(avg_words_in_response):.2f}")
print(f"Max Words in Response: {max(max_words_in_response)}")
print(f"Avg Words per Turn: {sum(avg_words_per_turn) / len(avg_words_per_turn):.2f}")
print(f"Max Words per Turn: {max(max_words_per_turn)}")
```
