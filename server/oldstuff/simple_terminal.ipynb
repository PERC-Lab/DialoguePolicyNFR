```python
import time
import select
import sys

def read_outputs(process, timeout=2.0):
    output = []
    last_data_time = time.time()
    fd = process.stdout.fileno()
    while True:
        ready, _, _ = select.select([fd], [], [], 0.1) # waits 0.1 second to see if there is any output
        if ready:
            line = process.stdout.readline()
            if line:
                output.append(line)
                last_data_time = time.time()
                sys.stdout.write(line)
                sys.stdout.flush()
        else:
            if time.time() - last_data_time > timeout:
                break

    return "".join(output)
#process.stdin.write("pwd\n")
#process.stdin.flush()
```
