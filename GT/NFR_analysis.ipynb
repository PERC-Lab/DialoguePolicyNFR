```python
import json
with open('NFR_balanced3.json', 'r') as file:
        data = json.load(file)
```

```python
data1 = []
i = 0
temp = []
for nfr in data:
    i += 1
    if not nfr.get('satisfaction_level'):
        print(i, nfr['id'])
    nfr['legacy_id'] = nfr['id']
    del nfr['new_id']
    nfr['id'] = i
    temp.append(nfr)
    if i%10 == 0 and i > 5:
        data1.append(temp)
        temp = []


```

```python
with open('NFR_balancedt4.json', 'w') as f:
    json.dump(data1, f, indent=4, ensure_ascii=False)
```
