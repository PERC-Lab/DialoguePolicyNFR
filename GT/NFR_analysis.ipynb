```python
import json
with open('NFR_balanced.json', 'r') as file:
        data = json.load(file)
```

```python
data1 = []
i = 0
for b in data:
    for nfr in b:
        i += 1
        nfr['new_id'] = i
        data1.append(nfr)
        if not nfr.get("satisfaction_level") or not nfr.get("reasoning"):
            if nfr['id'] <=40 :
                continue
            else:
                print(nfr['id'] )
```

```python
with open('NFR_balanced1.json', 'w') as f:
    json.dump(data1, f, indent=4, ensure_ascii=False)
```
