```python
import json

with open("data/conversations.json", "r", encoding="utf-8") as f:
    conversations = json.load(f)

with open("data/nfr_responses.json", "r", encoding="utf-8") as f:
    nfr_responses = json.load(f)
```

```python
for pid, responses in nfr_responses.items():
    for response in responses:
        response["satisfaction_level"] = ""
        response["reasoning"] = ""
        response["code_location"] = []
        response["comp"] = 0
```

```python
for pid, responses in conversations.items():
    for response in responses: #Recollections, Expansion, Refinement, Follow-up, UI, UR, IR, Error, HM, GD, RD, AR
        response["nfr_id"] = []
        response["Recollections"] = ""
        response["Expansion"] = ""
        response["Refinement"] = ""
        response["Follow-up"] = ""
        response["UI"] = ""
        response["UR"] = ""
        response["IR"] = ""
        response["Error"] = ""
        response["HM"] = ""
        response["GD"] = ""
        response["RD"] = ""
        response["AR"] = ""
        # "Context Memory": "Y",
        #"Self-correction": "",
        #"Self-affirmation": "",
        #"Proactive Interaction": "",
        #"Instruction Clarification": ""
        
```

```python
with open("annotated_data/nfr_responses.json", "w", encoding="utf-8") as f:
    json.dump(nfr_responses, f, indent=4)
```

```python
with open("annotated_data/conversations.json", "w", encoding="utf-8") as f:
    json.dump(conversations, f, indent=4)
```
